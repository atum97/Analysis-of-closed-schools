## 독립변수 산점도

```python
df = pd.read_excel('폐교데이터_전처리_완성.xlsx')
df.head()
```

```python
df.columns
```

```python
df = pd.concat([df['법정동명_시군구'],df['법정동명_읍면동'],df['폐교명'],df['건물연면적'],df['대지규모'],df['건물대장가격'],df['대지대장가격'],df['총가격']],
         axis=1, ignore_index=False)
df
```

```python
df.isna().sum()
```

## 결측값 제거 함수

```python
for col in df.loc[:,"건물연면적":].columns:
    df[col].fillna(0, inplace = True)
```

## 산점도

```python
df_mean = df.groupby(by='법정동명_시군구', as_index = False).mean()
df_mean
```

```python
import math
import numpy as np
```

```python
df_mean.loc[df_mean['건물대장가격'].isna()]
```

```python
x_val = df_mean['대지대장가격'].apply(np.log)
y_val = df_mean['건물대장가격'].apply(np.log)
```

```python
sns.scatterplot(x=x_val, y=y_val, hue=df_mean['법정동명_시군구']).get_legend().remove()
```

```python
sns.scatterplot(x=x_val, y=y_val, hue=df_mean['법정동명_시군구'])
```
