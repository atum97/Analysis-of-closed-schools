```python
df = pd.read_csv('mianhae2.CSV', encoding='euc-kr')
df.head()
```

```python
df.columns
```

```python
df_new = pd.concat([df['임의고유번'],df.loc[:,'학교_cnt':'보건_cnt'], df.loc[:,'법원_cnt':'국세_cnt'], df.loc[:,'고용_cnt':'tour_cnt'],df['pop_sum_sum']], axis=1, ignore_index=False)
df_new.head()
```

```python
df_new.to_csv('cnt_전처리_최종.csv', encoding='euc-kr', index=False)
```

```python
df_new.corr().head()
```

```python
plt.title('Count_Correlation', fontsize=20)
color_map = plt.cm.jet
plt.rcParams['figure.figsize']=(10,10)
sns.heatmap(df_new.corr(),
           annot=True,
           linewidth=0.05,
           vmax=1.0,
           square=True,
           cmap=color_map)
plt.show()
```

```python
color_map=plt.cm.jet
sns.heatmap(df_new.corr(),
           annot=True,
           linewidths=0.05,
           vmax=1.0,
           square=True,
           cmap = 'RdYlGn_r')
plt.show()
```
