## 컬럼추출, 컬럼별 연산 

```python
df_ingu = df.loc[:,"0~4세":"100세 이상"]
df_ingu.head()
```

```python
for col in df_ingu.columns:
    df_ingu[col] = df_ingu[col]/df['총인구']
    
df_ingu.head()    
```

```python
validation = 0
for i in range(len(df_ingu)):
    validation += df_ingu.iloc[i].sum()
    
print(validation, len(df_ingu))
```
