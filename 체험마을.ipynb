## 도로명 읍면동 분리

```python
def cut (lst):
    return lst[0:len(lst)-2]

def attach (lst):
    return " ".join(lst)
```

```python
df_cut = df['소재지도로명주소'].str.strip()
df_cut = df_cut.str.split()
df_cut
```

```python
df_cut = df_cut.apply(cut).apply(attach)
df_cut
```

## code join 

```python
os.listdir()
```

```python
code = pd.read_excel('읍면동_법정코드.xlsx')
code.head()
```

```python
df_cut.head()
```

```python
df.head()
```

```python
df['소재지도로명주소'] = df_cut
df.head()
```

```python
df = pd.merge(df, code, left_on='소재지도로명주소', right_on='법정동명_읍면동', how='left')
df.head()
```

```python
df = pd.concat([df['법정동코드_읍면동'], df['소재지도로명주소'], df['체험마을명'], df['체험프로그램구분'], df['위도'], df['경도']],
         axis=1, ignore_index=False)
df
```

```python
df.to_csv('체험마을1.csv', encoding='euc-kr')
```

## 체험마을 개수 카운팅 / 결합 

```python
df_count = df.groupby(by='법정동명_읍면동', as_index=None).count()
df_count 
```

```python
df_count = df_count.drop(['위도', '경도', '체험프로그램구분', '법정동코드_읍면동_y'], axis=1)
df_count
```

```python
df= pd.merge(df, df_count, left_on='법정동명_읍면동', right_on='법정동명_읍면동', how='left')
df
```

```python
df.rename(columns={'법정동코드_읍면동_y':'법정동코드_읍면동',
                  '체험마을명_x':'체험마을명',
                  '체험마을':'체험마을 개수'}, inplace=True)
```
